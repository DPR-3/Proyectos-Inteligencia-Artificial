```python
# Comienzo del dataset
resumeDataSet = pd.read_csv("UpdatedResumeDataSet.csv")
resumeDataSet['cleaned_resume'] = ' '
resumeDataSet.head()
```

```python
# Info del dataset
resumeDataSet.info()
```

```python
# Mostramos las distintas categorias del curriculum
print ("Categorias de curriculums:\n")
print (resumeDataSet['Category'].unique())
```

```python
# Mostramos las Categorias y el numero de registros de cada Categoria
print ("Mostramos las Categorias y el numero de registros de cada Categoría: \n ")
print (resumeDataSet['Category'].value_counts())
```

```python
# Grafica de las categorias y registros
import seaborn as sns 
plt.figure(figsize=(20,5))
plt.xticks(rotation=90)
ax=sns.countplot(x="Category", data=resumeDataSet)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
    plt.grid()
```
